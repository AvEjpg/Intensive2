```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('train.csv')
df.shape
df.head(10)
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(df.isnull(), cmap='BuPu_r')
```

```python
threshold = 0.7
df_c = df.loc[:, df.isnull().mean() < threshold]
df_c.head(10)
```

```python
threshold = 0.7
df_c = df_c[df_c.isnull().mean(axis=1) < threshold]
df_c.shape
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(df_c.isnull(), cmap='BuPu_r')
```

```python
# Подсчет количества непустых значений в каждой строке
df_c['non_null_count'] = df_c.apply(lambda row: row.count(), axis=1)
# Сортировка по количеству непустых значений в порядке убывания
df_s = df_c.sort_values(by='non_null_count', ascending=False)
# Оставляем только 5000 самых заполненных строк
df_s = df_s.head(5000)
# Удаляем временный столбец 'non_null_count'
df_s = df_s.drop(columns=['non_null_count'])
```

```python
df_s.shape
df_s.head(10)
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(df_s.isnull(), cmap='BuPu_r')
```

```python
# Заполнение медианой для всех числовых столбцов с пропусками
for col in df_s.select_dtypes(include=['float64', 'int64']):
    if df_s[col].isnull().sum() > 0:
        df_s[col] = df_s[col].fillna(df_s[col].median())
```

```python
non_numeric_columns = df_s.select_dtypes(exclude=['number'])
print("Нечисловые столбцы:")
print(non_numeric_columns)
```

```python
df_s.drop(['col1454'], axis = 1, inplace = True)
df_s.drop(['report_date'], axis = 1, inplace = True)
```

```python
def del_duble(df_s):
  unique_columns = []
  for column in df_s.columns:
    if not any(df_s[column].equals(df_s[col]) for col in unique_columns):
      unique_columns.append(column)
  return df_s[unique_columns]

df_s = del_duble(df_s)
df_s.shape
```

```python
correlation_matrix = df_s.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('Матрица корреляций признаков')
plt.show()
```

```python
df_s.to_csv('cleardata.csv', index=False)
```
