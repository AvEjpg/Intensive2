```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('train.csv')
df.shape
df.head(10)
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(df.isnull(), cmap='BuPu_r')
```

```python
threshold = 0.7
df_c = df.loc[:, df.isnull().mean() < threshold]
df_c.head(10)
```

```python
threshold = 0.7
df_c = df_c[df_c.isnull().mean(axis=1) < threshold]
df_c.shape
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(df_c.isnull(), cmap='BuPu_r')
```

```python
# ÐŸÐ¾Ð´ÑÑ‡ÐµÑ‚ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð° Ð½ÐµÐ¿ÑƒÑÑ‚Ñ‹Ñ… Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ Ð² ÐºÐ°Ð¶Ð´Ð¾Ð¹ ÑÑ‚Ñ€Ð¾ÐºÐµ
df_c['non_null_count'] = df_c.apply(lambda row: row.count(), axis=1)
# Ð¡Ð¾Ñ€Ñ‚Ð¸Ñ€Ð¾Ð²ÐºÐ° Ð¿Ð¾ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ñƒ Ð½ÐµÐ¿ÑƒÑÑ‚Ñ‹Ñ… Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ Ð² Ð¿Ð¾Ñ€ÑÐ´ÐºÐµ ÑƒÐ±Ñ‹Ð²Ð°Ð½Ð¸Ñ
df_s = df_c.sort_values(by='non_null_count', ascending=False)
# ÐžÑÑ‚Ð°Ð²Ð»ÑÐµÐ¼ Ñ‚Ð¾Ð»ÑŒÐºÐ¾ 5000 ÑÐ°Ð¼Ñ‹Ñ… Ð·Ð°Ð¿Ð¾Ð»Ð½ÐµÐ½Ð½Ñ‹Ñ… ÑÑ‚Ñ€Ð¾Ðº
df_s = df_s.head(5000)
# Ð£Ð´Ð°Ð»ÑÐµÐ¼ Ð²Ñ€ÐµÐ¼ÐµÐ½Ð½Ñ‹Ð¹ ÑÑ‚Ð¾Ð»Ð±ÐµÑ† 'non_null_count'
df_s = df_s.drop(columns=['non_null_count'])
```

```python
df_s.shape
df_s.head(10)
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(df_s.isnull(), cmap='BuPu_r')
```

```python
# Ð—Ð°Ð¿Ð¾Ð»Ð½ÐµÐ½Ð¸Ðµ Ð¼ÐµÐ´Ð¸Ð°Ð½Ð¾Ð¹ Ð´Ð»Ñ Ð²ÑÐµÑ… Ñ‡Ð¸ÑÐ»Ð¾Ð²Ñ‹Ñ… ÑÑ‚Ð¾Ð»Ð±Ñ†Ð¾Ð² Ñ Ð¿Ñ€Ð¾Ð¿ÑƒÑÐºÐ°Ð¼Ð¸
for col in df_s.select_dtypes(include=['float64', 'int64']):
    if df_s[col].isnull().sum() > 0:
        df_s[col] = df_s[col].fillna(df_s[col].median())
```

```python
non_numeric_columns = df_s.select_dtypes(exclude=['number'])
print("ÐÐµÑ‡Ð¸ÑÐ»Ð¾Ð²Ñ‹Ðµ ÑÑ‚Ð¾Ð»Ð±Ñ†Ñ‹:")
print(non_numeric_columns)
```

```python
df_s.drop(['col1454'], axis = 1, inplace = True)
df_s.drop(['report_date'], axis = 1, inplace = True)
```

```python
def del_duble(df_s):
  unique_columns = []
  for column in df_s.columns:
    if not any(df_s[column].equals(df_s[col]) for col in unique_columns):
      unique_columns.append(column)
  return df_s[unique_columns]

df_s = del_duble(df_s)
df_s.shape
```

```python
correlation_matrix = df_s.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False)
plt.title('ÐœÐ°Ñ‚Ñ€Ð¸Ñ†Ð° ÐºÐ¾Ñ€Ñ€ÐµÐ»ÑÑ†Ð¸Ð¹ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¾Ð²')
plt.show()
```
